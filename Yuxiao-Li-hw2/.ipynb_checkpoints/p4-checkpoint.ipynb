## Multinomial Logistic Regression

```python
# theta_j: 1xd , xi: 1xd
def loss_function(theta,x,y):
    total_loss = 0
    n = np.shape(x)[0]

    for i in range(n):
        loss = 0
        denominator = sum([np.exp(theta[:,j].dot(x[i])) for j in range(k)])
    
        for l in range(k):
            if y[i] == l:
                loss = np.log(np.exp(theta[:,l].dot(x[i])) / denominator)  
        total_loss += loss
        
    return -(loss/n)
```

```python
def accuracy(theta, x, y):
    n = len(x)
    y_hat = theta.T.dot(x.T).T
    label = np.zeros(n)
    
    for i in range(n):
        label[i] = np.argmax(y_hat[i])
    
    accuracy = sum(label == y)/n
    
    return accuracy
```

```python
# grad: d x k

def grad(theta,x,p):
    n,d = np.shape(x)[0], np.shape(x)[1]
    temp_grad = np.zeros(d)
    
    for i in range(n):
        denominator = np.sum(np.exp(np.dot(theta.T,x[i])))
        for l in range(k):
            theta_t = theta[:,p]
            # y: train_y
            if y[i] == l:
                if l == p:
                    temp_grad += - x[i] + x[i] * (np.exp(theta_t.dot(x[i])/denominator))
                else:
                    temp_grad += x[i] * (np.exp(theta_t.dot(x[i])/denominator))
    
    temp_grad =  temp_grad/n
    # print(np.linalg.norm(temp_grad))
    return temp_grad                


def allgrad(theta,x):
    n,d = np.shape(x)[0], np.shape(x)[1]
    gradient = np.zeros([d,k])
    
    for p in range(k):
        gradient[:,p] = grad(theta,x,p)
        
    return gradient
```

```python
def AGD(alpha, T, x, y, theta0):
    theta1 = theta0 - alpha * allgrad(theta0, x)

    train_loss = []
    test_loss = []
    # grad_list = []
    
    for t in range(1,T):
        yt = theta1 + (t-1)/(t+2) * (theta1 - theta0)
        temp_grad = allgrad(theta1, yt.T)
        theta_t = yt - alpha * temp_grad
        theta0 = theta1
        theta1 = theta_t
        # grad_list.append(np.linalg.norm(temp_grad))
        
        # calcula the training loss and test loss
        train_loss.append( loss_function(theta_t, x, y) )
        test_loss.append( loss_function(theta_t, test_x, test_y) )
        
    return theta_t, train_loss, test_loss
```

 ## Ordinal Logistic Regression

```python
# sigmoid function : t nxk
def h(t):
    return 1/(1+np.exp(-t))

# dh(t)/dt
def dh(t):
    return np.exp(-t)/(1+np.exp(-t))**2

# construct A nXk : if y[i] == j,1,0
def a(x,y):
    n = len(x)
    a = np.zeros([n,k]) 
    for i in range(n):
        for j in range(k):
            if y[i] == j:
                a[i,j] = 1 
                
    return a


# construct M: kXk
M = np.diag(np.ones(k,))
for i in range(k):
    for j in range(k-1):
        if i == j:
            M[i,j+1] = -1 
```

```python
# theta:dx1, z:kX1, x:dxk -> scaler
def loss_function2(theta,Z,x,y):
    n = len(x)
    A = a(x,y)
    temp = (x @ theta @ np.ones((1,k))) - (np.ones((n,1)) @ Z.T)
    hx = h(temp)
    # print(hx.all == 0)
    hm = hx @ M 
    loss = A * np.where(hm <= 0, hm, np.log(hm)) # nxk

    return (-1/n) * np.ones((1,n)) @ loss @ np.ones((k,1))



def accuracy2(theta, Z, x, y):
    n = len(x)
    sigmod =  h(x @ theta @ np.ones((1,k)) - np.ones((n,1)) @ Z.T)
    accuracy = (np.argmax(sigmod @ M, axis =1) == y ).sum() /n
    
    return accuracy
```

```python
def grad2(theta, Z, x, y):
    A = a(x,y)
    n = len(x)
    hx = h( (x @ theta @ np.ones((1,k))) - (np.ones((n,1)) @ Z.T)) 
    Qa = np.diag(A.flatten("F"))                     # nk * nk
    hm = (hx @ M).flatten("F")                       # nk * nk
    Qlog = np.diag( np.where( hm == 0, 0, 1/hm) )    # nk * nk
    Qh = np.diag(dh(x @ theta @ np.ones((1,k)) - np.ones((n,1)) @ Z.T).flatten("F"))  # nk * nk
    
    # temp: 1 x nk
    temp = (np.kron(np.ones((1,k)), np.ones((1,n)))) @ Qa @ Qlog @ np.kron(M.T,np.eye(n)) @ Qh
    temp_theta = np.kron(np.ones((k,1)), x)      # nk * d
    temp_z = np.kron(np.ones((n,1)), np.eye(k))  # nk * k
    
    grad_theta = (-1/n) * temp @ temp_theta        # 1 * d
    grad_z = (1/n) * temp @ temp_z                 # 1 * k

    return grad_theta, grad_z
```

```python
def AGD2(alpha, T, x, y, theta0, Z0):
    
    
    grad_theta1, grad_z1 = grad2(theta0, Z0, x, y)
    grad_theta1, grad_z1 = grad_theta1.reshape((d,1)), grad_z1.reshape((k,1))    
    # print("grad_theta1:"+ str(grad_theta1.shape))
    # print("grad_z1:"+ str(grad_z1.shape))
    theta1 = theta0 - alpha * grad_theta1
    Z1 = Z0 - alpha * grad_z1

#     print(theta1)
#     print(Z1)
    train_loss = []
    test_loss = []
    # grad_list = []
    
    for t in range(1,T):
        # learning theta and z
        y_theta = theta1 + (t-1)/(t+2) * (theta1 - theta0)
        y_Z = Z1 + (t-1)/(t+2) * (Z1 - Z0)
        
        theta_temp, z_temp = grad2(y_theta, y_Z, x, y)
        theta_temp, z_temp = theta_temp.reshape((d,1)), z_temp.reshape((k,1))
        theta_t = y_theta - alpha * theta_temp
        Zt = y_Z - alpha * z_temp
        
        # update theta and z
        theta0, Z0 = theta1, Z1
        theta1, Z1 = theta_t, Zt
        # grad_list.append(np.linalg.norm(temp_grad))
 
          
        # calcula the training loss and test loss
        train_loss_t = loss_function2(theta_t.reshape((d,1)), Zt.reshape((k,1)), x, y)[0,0] 
        train_loss.append(train_loss_t)
        test_loss.append( loss_function2(theta_t.reshape((d,1)), Zt.reshape((k,1)), test_x, test_y)[0,0] )
        
        if t % 100 == 0:
            print("Cost after  {} iteration: train loss is {}".format(t, train_loss_t))


    return theta_t, Zt, train_loss, test_loss
```

```python
# initialization
theta2 = np.zeros([d,1])
Z = np.array([[3],[2],[1],[0]])
alpha = 0.01
T = 2000
```

```python
theta_hat2, z_hat, train_loss2,test_loss2 = AGD2(alpha, T, x, y, theta2, Z)
theta_hat2, z_hat
```

```python
plt.xlabel('iteration time')
plt.ylabel('loss')
plt.plot(range(1,T),train_loss2,linewidth = '1',color='red',label='train loss')
plt.plot(range(1,T),test_loss2,linewidth = '1',color='blue',label='test loss')
plt.legend()
plt.show()
```
